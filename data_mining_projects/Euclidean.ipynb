```python
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances
```

```python
# Load your heart dataset
df_heart = pd.read_csv('heart.csv')
# Extract relevant features (attributes) for computing Euclidean distance
attributes = ['age', 'trestbps', 'chol']
df_heart
```

```python
# Extract attribute values for the dataset
data_heart = df_heart[attributes].values
data_heart
```

```python
# Compute Euclidean distances between data points
distances_heart = euclidean_distances(data_heart)
```

```python

similarity_matrix_heart = 1 / (1 + distances_heart)
```

```python

similarity_df_heart = pd.DataFrame(similarity_matrix_heart, index=df_heart.index, columns=df_heart.index)
similarity_df_heart
```

```python
similar_instances = similarity_df_heart.loc[17].sort_values(ascending=False)

# Print the top similar instances
print(similar_instances.head(30))
```
